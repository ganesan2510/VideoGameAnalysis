```python
df_games.dropna(inplace = True)
df_sales.dropna(inplace = True)
df_games.head()
```

```python
df_games["Team"] = (
    df_games["Team"]
    .fillna("")
    .astype(str)
    .str.replace(r'[\[\],\'"]', "", regex=True)
    .str.strip()
)
df_games["Genres"] = (
    df_games["Genres"]
    .fillna("")
    .astype(str)
    .str.replace(r'[\[\],\'"]', "", regex=True)
    .str.strip()
)
df_games["Reviews"] = (
    df_games["Reviews"]
    .fillna("")
    .astype(str)
    .str.replace(r'[\[\],\'"]', "", regex=True)
    .str.strip()
)
df_games.head()
```

```python
# Clean column names
# -----------------------------
df_games.columns = (
    df_games.columns
    .str.strip()              # remove leading/trailing spaces
    .str.lower()              # lowercase
    .str.replace(" ", "_")    # replace spaces with underscores
)
```

```python
#Remove unwanted index column (if exists)
df_games.drop(columns=["unnamed:_0"])
```

```python
df_games.columns
```

```python
df_games['times_listed'] = df_games['times_listed'].fillna(0)
df_games['playing'] = df_games['playing'].fillna(0)
df_games["release_date"] = (
    pd.to_datetime(df_games["release_date"], errors="coerce")
    .dt.strftime("%Y-%m-%d")
)
df_games['times_listed']
```

```python
def convert_k(value):
    if pd.isna(value):
        return 0
    #value = str(value).strip().upper()
    if value.endswith("K"):
        return float(value.replace("K", "")) * 1000
    else:
        return int(float(value))
df_games["times_listed"] = df_games["times_listed"].apply(convert_k).astype("Int64")
df_games["plays"] = df_games["plays"].apply(convert_k).astype("Int64")
df_games["playing"] = df_games["playing"].apply(convert_k).astype("Int64")
df_games["backlogs"] = df_games["backlogs"].apply(convert_k).astype("Int64")
df_games["wishlist"] = df_games["wishlist"].apply(convert_k).astype("Int64")
df_games["number_of_reviews"] = df_games["number_of_reviews"].apply(convert_k).astype("Int64")

df_games.head()
```
